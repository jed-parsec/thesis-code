# MERGING DEAUTH CSV DATASET

```python
import pandas as pd
import os
from tqdm.notebook import tqdm  # Use tqdm.notebook for Jupyter-specific progress bar

# Specify the path to your Deauth folder
folder_path = '1.Deauth'  # Adjust this path if needed (e.g., './Deauth' or full path)

# Create an empty list to store dataframes
all_dataframes = []

# Get list of CSV files
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Loop through all CSV files with progress bar
for filename in tqdm(csv_files, desc="Loading CSV files"):
    file_path = os.path.join(folder_path, filename)
    try:
        # Read each CSV file and append to the list
        df = pd.read_csv(file_path)
        all_dataframes.append(df)
        # print(f"Successfully loaded: {filename}")  # Optional, commented out since tqdm shows progress
    except Exception as e:
        print(f"Error loading {filename}: {str(e)}")

# Concatenate all dataframes into a single dataset
if all_dataframes:  # Check if there are any dataframes to merge
    merged_dataset = pd.concat(all_dataframes, ignore_index=True)
    print("\nDataset merged successfully!")
    print(f"Total rows: {merged_dataset.shape[0]}")
    print(f"Total columns: {merged_dataset.shape[1]}")
    
    # Optional: Display the first few rows
    print("\nFirst 5 rows of the merged dataset:")
    print(merged_dataset.head())
    
    # Optional: Save the merged dataset to a new CSV file
    output_file = 'deauth.csv'
    merged_dataset.to_csv(output_file, index=False)
    print(f"\nMerged dataset saved as '{output_file}'")
else:
    print("No CSV files found in the specified folder or failed to load any files.")

# Now you can use 'merged_dataset' for further analysis
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Load the dataset
X = pd.read_csv("X_test.csv")  # Features
y = pd.read_csv("y_test.csv").values.ravel()  # Labels (flattened to 1D)

# Step 2: Combine X and y into a dataframe for resampling
df = X.copy()
df['label'] = y  # Add labels to the dataset

# Step 3: Check class distribution
class_counts = df['label'].value_counts()
print(f"Original Class Distribution:\n{class_counts}\n")

# Step 4: Undersample the "Normal" traffic
df_deauth = df[df['label'] == 'Deauth']  # Keep all "Deauth" samples
df_normal = df[df['label'] == 'Normal'].sample(n=13000, random_state=42)  # Downsample "Normal" to 13,000 samples

# Step 5: Combine the balanced dataset
df_balanced = pd.concat([df_deauth, df_normal])
X_balanced = df_balanced.drop(columns=['label'])
y_balanced = df_balanced['label']

# Step 6: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.3, random_state=42)

# Step 7: Train a Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Step 8: Make predictions
y_pred = rf_model.predict(X_test)

# Step 9: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:\n", report)

# Step 10: Plot Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=set(y_test), yticklabels=set(y_test))
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix (After Undersampling)")
plt.show()
```

# MERGING ALL CSV: deauth, evil_twin, krack, rogue_ap

```python
import pandas as pd
from tqdm import tqdm

# Define dataset filenames
datasets = ["deauth.csv", "evil_twin.csv", "krack.csv", "rogue_ap.csv"]
df_list = []  # List to store downsampled DataFrames
NORMAL_LIMIT = 50000  # Limit "Normal" traffic to 50,000

# Process each dataset with a progress bar
print("[*] Processing datasets...")
for file in tqdm(datasets, desc="Processing"):
    df = pd.read_csv(file)

    # Check if "label" column exists
    if 'Label' not in df.columns:
        print(f"[!] Warning: 'label' column not found in {file}. Skipping...")
        continue

    # Downsample "Normal" class if necessary
    df_normal = df[df['Label'] == 'Normal']
    df_attack = df[df['Label'] != 'Normal']

    if len(df_normal) > NORMAL_LIMIT:
        df_normal = df_normal.sample(n=NORMAL_LIMIT, random_state=42)

    # Combine "Normal" and attack data
    df_balanced = pd.concat([df_normal, df_attack], ignore_index=True)
    df_list.append(df_balanced)

# Merge all datasets into one
df_merged = pd.concat(df_list, ignore_index=True)

# Display final class distribution
print("\n[*] Final class distribution:")
print(df_merged['label'].value_counts())

# Save merged dataset
df_merged.to_csv("awid3_v1_downsampled.csv", index=False)
print("[] Downsampling & merging completed! Saved as 'awid3_v1_downsampled.csv'.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# Load dataset
print("[*] Loading awid3_v1_downsampled.csv...")
awid_data = pd.read_csv("awid3_v1_downsampled.csv")

# Identify the correct label column name
possible_labels = ["Label", "label", "class", "attack_type"]
target_column = next((col for col in possible_labels if col in awid_data.columns), None)

if target_column is None:
    raise ValueError(" Could not find the target label column in dataset!")

print(f"[*] Using target column: {target_column}")

# Drop rows where the label is '0'
print("[*] Dropping rows where the label is '0'...")
awid_data = awid_data[awid_data[target_column] != '0']

# Load column order from v_X_test.csv
v_X_test = pd.read_csv("v_X_test.csv", nrows=1)  # Load only header row
expected_columns = list(v_X_test.columns)  # Preserve the exact order

# Keep only the columns present in v_X_test (excluding target column)
print("[*] Dropping extra columns and aligning order...")
filtered_columns = [col for col in expected_columns if col in awid_data.columns]
awid_data = awid_data[filtered_columns + [target_column]]

# Drop columns with >90% missing values
print("[*] Removing columns with >90% missing values...")
threshold = 0.9 * len(awid_data)
awid_data = awid_data.dropna(thresh=threshold, axis=1)

# Convert non-numeric columns (EXCEPT the target column)
print("[*] Converting non-numeric columns to numeric (excluding target labels)...")
for col in tqdm(awid_data.columns):
    if col != target_column and awid_data[col].dtype == "object":
        awid_data[col] = pd.factorize(awid_data[col])[0]

# Replace NaN values with 0
print("[*] Replacing NaN values with 0...")
awid_data = awid_data.fillna(0)

# Split dataset into features (X) and labels (y)
print("[*] Splitting dataset into features and labels...")
X = awid_data.drop(columns=[target_column])  # Features
y = awid_data[target_column].astype(str)  # Labels (categorical)

# Train-test split (Stratified to maintain label distribution)
print("[*] Splitting dataset into train and test sets...")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)

# Reorder X_train and X_test to match v_X_test.csv column arrangement
print("[*] Reordering columns to match v_X_test.csv...")
X_train = X_train.reindex(columns=expected_columns, fill_value=0)
X_test = X_test.reindex(columns=expected_columns, fill_value=0)

# Save processed datasets
print("[*] Saving processed datasets...")
X_train.to_csv("preprocessed_X_train.csv", index=False)
X_test.to_csv("preprocessed_X_test.csv", index=False)
y_train.to_csv("preprocessed_y_train.csv", index=False)
y_test.to_csv("preprocessed_y_test.csv", index=False)

print("[] Preprocessing complete! Rows with label '0' have been removed, and column order matches v_X_test.csv.")
```

```python
import pandas as pd
import joblib
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from catboost import CatBoostClassifier
from sklearn.preprocessing import LabelEncoder

# Step 1: Load Pre-trained Base Models
catboost_model = joblib.load("BoostingClassifier_catboost_model.pkl")
gradient_boosting_model = joblib.load("BoostingClassifier_gradient_boosting_model.pkl")

print("[+] Base models loaded successfully!")

# Step 2: Load the Preprocessed Test Dataset
print("[*] Loading preprocessed test dataset...")
X_test = pd.read_csv("preprocessed_X_train.csv")  
y_test = pd.read_csv("preprocessed_y_train.csv")  

# Convert y_test to 1D array
y_test = y_test.values.ravel()  

# Encode labels (if needed)
label_encoder = LabelEncoder()
y_test = label_encoder.fit_transform(y_test)  

print("[+] Test dataset loaded!")

# Step 3: Generate Base Model Predictions
print("[*] Generating base model predictions...")
catboost_preds = catboost_model.predict(X_test)
gradient_boosting_preds = gradient_boosting_model.predict(X_test)

# Convert predictions to numpy arrays and ensure they are 1D
catboost_preds = np.array(catboost_preds).ravel()
gradient_boosting_preds = np.array(gradient_boosting_preds).ravel()

# Convert predictions to encoded format if they are categorical
if isinstance(catboost_preds[0], str):
    catboost_preds = label_encoder.transform(catboost_preds)

if isinstance(gradient_boosting_preds[0], str):
    gradient_boosting_preds = label_encoder.transform(gradient_boosting_preds)

# Combine predictions into new dataset for meta-model
stacked_features = pd.DataFrame({
    "catboost_pred": catboost_preds,
    "gradient_boosting_pred": gradient_boosting_preds
})

print("[+] Stacked features created!")

# Step 4: Train the Meta Model (CatBoost)
meta_model = CatBoostClassifier(iterations=100, depth=6, learning_rate=0.1, verbose=0)
meta_model.fit(stacked_features, y_test)  

print("[+] Meta classifier trained!")

# Step 5: Make Final Predictions using Meta Model
print("[*] Making final predictions...")
y_pred = meta_model.predict(stacked_features)

# Ensure y_pred is also encoded properly
if isinstance(y_pred[0], str):
    y_pred = label_encoder.transform(y_pred)

print("[+] Final predictions completed!")

# Step 6: Evaluate Model Performance
print("\n[*] Evaluating Model Performance...")

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, zero_division=1)  
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"\nAccuracy: {accuracy}")
print("\nClassification Report:\n", report)

# Step 7: Check Class Distribution
unique, counts = np.unique(y_test, return_counts=True)
print("Class distribution in y_test:", dict(zip(unique, counts)))

# Step 8: Plot Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()

print("[+] Stacked model evaluation complete!")
```
